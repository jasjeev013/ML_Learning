```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + 1.5 * X + 2 + np.random.randn(100, 1)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler();
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual Values', marker='o')
plt.plot(y_pred, label='Predicted Values', marker='x')
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print('Mean Squared Error:',mse)
print('R2 Score:',r2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, regression.predict(X_train), color='red', linewidth=2, label='Best fit line')
plt.title('Scatter plot of Training Data with Best Fit Line')
plt.xlabel('X_train')
plt.ylabel('y_train')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train_poly,y_train)
y_pred = regression.predict(X_test_poly)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual Values', marker='o')
plt.plot(y_pred, label='Predicted Values', marker='x')
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y_test,y_pred)
print('Mean Squared Error:',mse)
r2 = r2_score(y_test,y_pred)
print('R2 Score:',r2)
```

```python
# Scatter plot of all points
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')

# Generate a range of values for X to plot the best fit curve
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)

# Predict y values using the polynomial regression model
y_range_pred = regression.predict(X_range_poly)

# Plot the best fit curve
plt.plot(X_range, y_range_pred, color='red', linewidth=2, label='Best fit curve')

plt.title('Scatter plot of all points with Best Fit Curve')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
