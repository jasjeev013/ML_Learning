```python
import pandas as pd
messages = pd.read_csv('spam.csv', encoding='latin-1')
messages = messages.drop(labels = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis = 1)
messages.columns = ["label", "text"]
messages['label'] = messages['label'].map({'ham': 0, 'spam': 1})
print(messages.head())
```

```python
import re
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
# from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
import re

lemmatizer = WordNetLemmatizer()

corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['text'][i])  # Remove special characters
    review = review.lower()  # Convert to lowercase
    review = review.split()  # Tokenize
    review = [lemmatizer.lemmatize(word) for word in review]  # Lemmatization
    review = ' '.join(review)  # Reconstruct the cleaned text
    corpus.append(review)


# Check the first few cleaned messages
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(words)
```

```python
import numpy as np
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv], axis=0)
```

```python
from tqdm import tqdm
```

```python
X = []
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```
