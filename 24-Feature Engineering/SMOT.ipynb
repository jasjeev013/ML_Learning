```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=2,n_clusters_per_class=1, weights=[0.90], random_state=12,n_redundant=0)
```

```python
import pandas as pd 
```

```python
df1 = pd.DataFrame(x, columns=['feature1','feature2'])
df2= pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1,df2], axis=1)
```

```python
final_df
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature1'], final_df['feature2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y = oversample.fit_resample(final_df[['feature1','feature2']], final_df['target'])
```

```python
df1 = pd.DataFrame(x, columns=['feature1','feature2'])
df2= pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1,df2], axis=1)
```

```python
plt.scatter(final_df['feature1'], final_df['feature2'], c=final_df['target'])
```
