```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.displot(df['age'],kde=True)
```

```python
df['Age_median'] = df['age'].fillna(df['age'].mean())
```

```python
df[['Age_median','age']]
```

```python
sns.displot(df['Age_median'],kde=True)
sns.displot(df['age'],kde=True)
```

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
sns.displot(df['Age_median'],kde=True)
sns.displot(df['age'],kde=True)
```
